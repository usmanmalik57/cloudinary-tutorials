## Importing Required Libraries

```python
from dotenv import load_dotenv
import os
import glob
import cloudinary
import cloudinary.uploader
from cloudinary import CloudinaryVideo

load_dotenv()

cloudinary.config( 
  cloud_name = os.getenv("CLOUD_NAME"), 
  api_key = os.getenv("API_KEY"), 
  api_secret = os.getenv("API_SECRET")
)
```

## Optimizing Video Bitrate

### Uploading and Optimizing Bitrates in two Steps

```python
## Video link : https://pixabay.com/videos/basketball-hoop-net-urban-street-208442/

response = cloudinary.uploader.upload("../video-samples/208442_tiny.mp4", 
  resource_type = "video",
  public_id = "basketball"
)
```

```python
response = CloudinaryVideo("basketball").video(bit_rate="250k")
```

```python
cloudinary.utils.cloudinary_url("basketball", streaming_profile="auto", resource_type="video")
```

### Setting Bitrate during Video Upload

```python
cloudinary.uploader.upload("../video-samples/208442_tiny.mp4", 
  resource_type = "video",
  public_id = "basketball",
  transformation = {"bit_rate": "128k"}
)
```

### Setting Bitrates for Multiple Videos

```python

folder = "../video-samples"
video_paths = glob.glob(os.path.join(folder, "*.mp4"))

for file_path in video_paths:
    public_id = os.path.splitext(os.path.basename(file_path))[0]

    print(f"Uploading {file_path} as {public_id}...")

    cloudinary.uploader.upload(file_path, 
    resource_type = "video",
    public_id = public_id,
    transformation = {"bit_rate": "128k"}
    )       

print("All uploads completed!")
```
