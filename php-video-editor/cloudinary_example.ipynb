## Importing Required Libraries

```python
from dotenv import load_dotenv
import os
import glob
import cloudinary
import cloudinary.uploader
from cloudinary import CloudinaryVideo

load_dotenv()

cloudinary.config( 
  cloud_name = os.getenv("CLOUD_NAME"), 
  api_key = os.getenv("API_KEY"), 
  api_secret = os.getenv("API_SECRET")
)
```

## Optimizing Single Video

```python
response = cloudinary.uploader.upload("sample.mp4", 
  resource_type = "video",
  public_id = "rodents_video"
)

response = CloudinaryVideo("rodents_video").video(width=512, crop="scale", quality=10)
response
```

## Optimizing Multiple Videos

```python

video_paths = ['sample.mp4', 
               'sample2.mp4']

for file_path in video_paths:
    public_id = os.path.splitext(os.path.basename(file_path))[0]

    print(f"Uploading {file_path} as {public_id}...")

    response = cloudinary.uploader.upload(
        file_path,
        resource_type="video",
        public_id=public_id,
        transformation={"width": 512, 
                        "crop": "scale",
                        "quality": 30}
    )

    print(f"Upload response for {public_id}: {response}")

print("All uploads completed!")
```
